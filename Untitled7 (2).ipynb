```python
# Calculate the distribution of employees across each team
team_distribution = data['Team'].value_counts()

# Calculate the percentage split
team_percentage = team_distribution / team_distribution.sum() * 100

# Combine into a single DataFrame for better visualization
team_distribution_df = pd.DataFrame({
    'Team': team_distribution.index,
    'Number of Employees': team_distribution.values,
    'Percentage': team_percentage.values
})

# Display the team distribution DataFrame
team_distribution_df.reset_index(drop=True, inplace=True)
team_distribution_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the size of the plot
plt.figure(figsize=(10,6))

# Create a bar plot for the number of employees per team
sns.barplot(x='Number of Employees', y='Team', data=team_distribution_df, palette='magma')

# Add title and labels
plt.title('Distribution of Employees Across Each Team')
plt.xlabel('Number of Employees')
plt.ylabel('Team')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#Teams like the New Orleans Pelicans, Memphis Grizzlies, Utah Jazz, New York Knicks, and Milwaukee Bucks have the highest number of employees, each having around 15-19 employees.
#The percentage of employees across these teams ranges from about 3% to 4%
```
